## Pandas, Scatter Plots

```python
# Generate scatter plots to explore relationship between different factors
recent_grads.plot(x='Sample_size', y='Median',
                  kind='scatter',
                  title='Sample_size vs. Median')

recent_grads.plot(x='Sample_size', y='Unemployment_rate',
                  kind='scatter',
                  title='Sample_size vs. Unemployment_rate')

recent_grads.plot(x='Full_time', y='Median',
                  kind='scatter',
                  title='Full_time vs. Median')

recent_grads.plot(x='ShareWomen', y='Unemployment_rate',
                  kind='scatter',
                  title='ShareWomen vs. Unemployment_rate')

recent_grads.plot(x='Men', y='Median',
                  kind='scatter',
                  title='Men vs. Median')

recent_grads.plot(x='Women', y='Median',
                  kind='scatter',
                  title='Women vs. Median')

recent_grads.plot(x='ShareWomen', y='Median',
                  kind='scatter',
                  title='ShareWomen vs. Median')
```

Fromt the plot "Sample_size vs. Median", there are no obvious correlation between the two factors, we actually observe median earnings converge towards a baseline of \$40,000 as the sample size increase, hence we are able to infer that students in more popular majors (larger sample sizes) do not make more money. In fact, it would seem that graduates from some unusual majors are the ones that have the higher than usual median earnings.

Comparing the plots "Men vs. Median" and "Women vs. Median", we find that both the median value for both plots tend to converge as the student number increase, however, the convergence value for the first plot seems to be around \$40,000, but only around \$35,000 for the second, indicating that majors with more absolute number of female students earns less. This relationship is even better illustrated with the plot "ShareWomen vs. Median", where we clearly see a negative correlation between the two factors.

Looking at the plot "Full_time vs. Median", we see that there are no obvious correlation between number of full time employees and median salary for each major. The pattern is similar to that of the first plot, where a convergence can be observed as the sample size number grew, and majority of positive outliers are concentrated when the sample size is small instead.

## Pandas, Histograms

```python
# Generating histograms to explore different columns
ax = recent_grads['Sample_size'].hist(bins=25, range=(0,5000), grid=False)
ax.set_title("Sample_size")
```

```python
ax2 = recent_grads['Median'].hist(bins=20, range=(0, 100000), orientation='horizontal')
ax2.set_title("Median")
```

```python
ax3 = recent_grads['Employed'].hist(bins=30)
ax3.set_title("Employed")
```

```python
ax4 = recent_grads['Full_time'].hist(bins=15)
ax4.set_title("Full_time")
```

```python
ax5 = recent_grads['ShareWomen'].hist(bins=10, range=(0,1), cumulative=True)
ax5.set_title("ShareWomen Cumulative Hist")
```

```python
ax6 = recent_grads['Unemployment_rate'].hist(bins=15)
ax6.set_title("Unemployment_rate")
```

```python
ax7 = recent_grads['Men'].hist(bins=15)
ax7.set_title("Men")
```

```python
ax8 = recent_grads['Women'].hist(bins=15)
ax8.set_title("Women")
```

Answering the following questions:
- What percent of majors are predominantly male? Predominantly female?
    - From the cumulative histogram of "ShareWomen", we see that the 5th bar, which is the cumulative frequencies of all majors with less than 50% of female in its cohort, hence predominantly male majors, has a height of about 75 majors. If we take percentage of this over the 172 majors we are working with, we can conclude that about 43.6% of majors have a predominantly male student body, whilst 56.4% of majors are predominantly female.

- What's the most common median salary range?
    - From the horizontal histogram of "Median", we see that the most common median salary range is between \$30,000 and \$35,000.
